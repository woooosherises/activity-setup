```python
import requests
import time

def get_iss_location():
    url = "http://api.open-notify.org/iss-now.json"
    data_points = []

    for _ in range(100):
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            latitude = float(data['iss_position']['latitude'])
            longitude = float(data['iss_position']['longitude'])
            timestamp = time.time()

            data_point = {'latitude': latitude, 'longitude': longitude, 'timestamp': timestamp}
            data_points.append(data_point)
        else:
            print("Error occurred while retrieving ISS location.")

        # Sleep for 1 second between requests
        time.sleep(1)

    return data_points

# Usage:
iss_data = get_iss_location()
for data_point in iss_data:
    print(f"Latitude: {data_point['latitude']}, Longitude: {data_point['longitude']}, Timestamp: {data_point['timestamp']}")
```
